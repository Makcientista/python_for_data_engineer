```python
my_list = ["Iphone", "Samsung", "MI"]
my_list.insert(2, "Pixel") # inserir o elemento Pixel na posição do index 2nd 
print(my_list)
```

```python
# For: When you want to run loop without a condition
my_list = ["Iphone", "Samsung", "MI"]
for phone in my_list:
    print(phone)
```

```python
my_list = ["Iphone", "Samsung", "Pixel", "MI"]
i = 1
while i < len(my_list):
    print(my_list[i])
    i = i+1
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []
for fruit in fruits:
    if "a" in fruit:        
        newlist.append(fruit) # append para add as na nova lista (newlist) fruta que tem a letra a.
print(newlist)
```
